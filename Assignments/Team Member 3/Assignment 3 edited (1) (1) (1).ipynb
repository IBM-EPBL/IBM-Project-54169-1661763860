# replacing outliers

```python
Q1=df.Length.quantile(0.25)
Q3=df.Length.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
```

```python
df['Length']=np.where(df['Length']<lower_limit,0.545,df['Length'])
sns.boxplot(df.Length)
```

```python
Q1=df.Diameter.quantile(0.25)
Q3=df.Diameter.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Diameter']=np.where(df['Diameter']<lower_limit,0.425,df['Diameter'])
sns.boxplot(df.Diameter)
```

```python
Q1=df.Height.quantile(0.25)
Q3=df.Height.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Height']=np.where(df['Height']<lower_limit,0.1400,df['Height'])
df['Height']=np.where(df['Height']>upper_limit,0.1400,df['Height'])
sns.boxplot(df.Height)
```

```python
Q1=df.Wholeweight.quantile(0.25)
Q3=df.Wholeweight.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Wholeweight']=np.where(df['Wholeweight']>upper_limit,0.7995,df['Wholeweight'])
sns.boxplot(df.Wholeweight)
```

```python
Q1=df.Shuckedweight.quantile(0.25)
Q3=df.Shuckedweight.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Shuckedweight']=np.where(df['Shuckedweight']>upper_limit,0.3360,df['Shuckedweight'])
sns.boxplot(df.Shuckedweight)
```

```python
Q1=df.Visceraweight.quantile(0.25)
Q3=df.Visceraweight.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Visceraweight']=np.where(df['Visceraweight']>upper_limit,0.1710,df['Visceraweight'])
sns.boxplot(df.Visceraweight)
```

```python
Q1=df.Shellweight.quantile(0.25)
Q3=df.Shellweight.quantile(0.75)
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
df['Shellweight']=np.where(df['Shellweight']>upper_limit,0.2340,df['Shellweight'])
sns.boxplot(df.Shellweight)
```

# 8. x and y split

```python
# X---> Independent variable
X=df.drop(columns=['age'])
X.head()
```

```python
y=df.age
y.head()
```

# 9.Scaling the independent variable(X)

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
```

```python
scale_x=pd.DataFrame(scale.fit_transform(X),columns=X.columns)
scale_x.head()
```

# 10.train,test,split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scale_x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

# 11.Building the model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

# 12.Training the model

```python
model.fit(x_train,y_train)
```

# 13.Testing the model

```python
train_pred=model.predict(x_train)
pd.DataFrame({'actual trained y':y_train,'predicted trained y':train_pred})
```

```python
test_pred=model.predict(x_test)
pd.DataFrame({'actual tested y':y_test,'predicted tested y':test_pred})
```

# 14.Evaluation metrics

# a) Mean squared error

```python
from sklearn import metrics
metrics.mean_squared_error(y_test,test_pred)
```

# b) Root Mean Squared error

```python
np.sqrt(metrics.mean_squared_error(y_test,test_pred))
```

# c) R2 Score

```python
metrics.r2_score(y_train,train_pred)
```

```python
metrics.r2_score(y_test,test_pred)
```
